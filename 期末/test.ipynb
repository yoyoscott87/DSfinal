```python
import requests

def get_air_quality_index(component, value):
    if component == 'so2':
        if value <= 20:
            return 0
        elif value <= 80:
            return 1
        elif value <= 250:
            return 2
        elif value <= 350:
            return 3
        else:
            return 4
    elif component == 'no2':
        if value <= 40:
            return 0
        elif value <= 70:
            return 1
        elif value <= 150:
            return 2
        elif value <= 200:
            return 3
        else:
            return 4
    elif component == 'pm10':
        if value <= 20:
            return 0
        elif value <= 50:
            return 1
        elif value <= 100:
            return 2
        elif value <= 200:
            return 3
        else:
            return 4
    elif component == 'pm2_5':
        if value <= 10:
            return 0
        elif value <= 25:
            return 1
        elif value <= 50:
            return 2
        elif value <= 75:
            return 3
        else:
            return 4
    elif component == 'o3':
        if value <= 60:
            return 0
        elif value <= 100:
            return 1
        elif value <= 140:
            return 2
        elif value <= 180:
            return 3
        else:
            return 4
    elif component == 'co':
        if value <= 4400:
            return 0
        elif value <= 9400:
            return 1
        elif value <= 12400:
            return 2
        elif value <= 15400:
            return 3
        else:
            return 4
    
    else:
        return 'Unknown'

# 設定你的 API key 和經緯度
api_key = "b55132e7d0d008aa945b18e4cd2f5a02"  # 替換為你的 API key
lat = "39.9042"  # 替換為你想查詢的緯度
lon = "116.4074"  # 替換為你想查詢的經度

# 建立 API URL
url = f"http://api.openweathermap.org/data/2.5/air_pollution?lat={lat}&lon={lon}&appid={api_key}"

# 發送 GET 請求
response = requests.get(url)

# 檢查請求是否成功
if response.status_code == 200:
    # 以 JSON 格式列印回應內容
    data = response.json()
    
    # 獲取空氣污染數據
    components = data['list'][0]['components']
    
    # 定義各組成部分的權重
    weights = {'so2': 0.1, 'no2': 0.15, 'pm10': 0.2, 'pm2_5': 0.25, 'o3': 0.15, 'co': 0.15}
    
    # 計算綜合評分
    total_score = 0
    for component, value in components.items():
        if component in weights:
            aqi_score = get_air_quality_index(component, value)
            total_score += aqi_score * weights[component]
    
    print("綜合評分:", total_score)
else:
    print("Failed to retrieve data:", response.status_code)
```

```python
def map_score_to_level(score):
    if score <= 1:
        return "優秀"
    elif score <= 2:
        return "良好"
    elif score <= 3:
        return "一般"
    else:
        return "差"

# 計算綜合評分
total_score = 0
for component, value in components.items():
    if component in weights:
        aqi_score = get_air_quality_index(component, value)
        total_score += aqi_score * weights[component]

# 將評分映射到等級
air_quality_level = map_score_to_level(total_score)
print("空氣品質等級:", air_quality_level)
```

```python
import requests

def get_weather_by_coordinates(api_key, latitude, longitude):
    url = f"https://api.openweathermap.org/data/2.5/weather?lat={latitude}&lon={longitude}&appid={api_key}&units=metric"
    response = requests.get(url)
    response.raise_for_status()  # 如果请求失败，会抛出异常
    return response.json()

def get_tem(api_key, latitude, longitude):
    try:
        weather_data = get_weather_by_coordinates(api_key, latitude, longitude)
        temperature = weather_data["main"]["temp"]
        humidity = weather_data["main"]["humidity"]
        weather_description = weather_data["weather"][0]["description"]

        print(f"地點: {weather_data['name']}")
        print(f"溫度: {round(temperature, 2)} °C")
        print(f"濕度: {humidity} %")
        print(f"天氣: {weather_description}")

        return {
            "temperature": round(temperature, 2),
            "humidity": humidity,
            "weather_description": weather_description
        }
    except requests.exceptions.RequestException as e:
        print("API請求出錯：", e)
        return None
    except KeyError as e:
        print("JSON解析出錯：", e)
        return None
    except Exception as e:
        print("未知錯誤：", e)
        return None
latitude = 35.6895  # 例如：台北市的緯度
longitude = 139.6917 # 例如：台北市的經度
weather_info = get_tem(api_key, latitude, longitude)
```
